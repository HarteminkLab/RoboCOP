# Plot heatmaps for MNase-seq/ATAC-seq around annotated TSS, nucleosomes and TF binding sites

In this tutorial we will plot heatmaps similar to Figure 1 in the [RoboCOP paper](https://doi.org/10.1093/nar/gkab553).  They are similar to, rather than exactly the same, because in this notebook we use a subsampled set of MNase reads to speed things up.  To reproduce the exact plots in the paper, you will need to load the complete set of MNase reads.

```python
# include path to RoboCOP directory
import sys
sys.path.insert(0, '../pkg/')
from robocop.utils import plot_TSS_heatmap, plot_nuc_heatmap, plot_TF_heatmap
```

```python
# MNase-seq BAM file
bamFile = 'inputs/DM504_sacCer3_subset.bam'
# annotated nucleosome positions
nucleosomeFile = 'inputs/Chereji_2018_+1_-1_nucs.bed'
# annotated TF binding sites
tfFile = 'inputs/MacIsaac_sacCer3_liftOver_Abf1_Reb1.bed'
# annoted TSS file
tssFile = 'inputs/Park_2014_TSS.csv'
```

```python
# range for nucleosomal fragments
nucFrag = (127, 187)
# range for short fragments
shortFrag = (0, 80)
```

# Plot fragment midpoints around TF binding sites

```python
plot_TF_heatmap.plotTF(bamFile, tfFile, nucFrag, shortFrag, tf='ABF1')
```

# Plot fragment midpoints around annotated nucleosome dyads

```python
plot_nuc_heatmap.plotNuc(bamFile, nucleosomeFile, nucFrag, shortFrag)
```

# Plot fragment midpoints around annotated TSS

```python
plot_TSS_heatmap.plotTSS(bamFile, tssFile, nucFrag, shortFrag)
```
